```python
import pandas as pd
```

```python
df = pd.read_csv("referendum_107.csv", thousands=',')
df
```

```python
# chose only proposal 7
df7 = df.loc[df['公投案'] == "第7案"]
df7
```

```python
df7 = df7.drop(columns=["公投案","鄉鎮市區","投票率(%)","有效同意票數對投票權人數百分比(%)"])
```

```python
type(df.iloc[2,4])
```

```python
df7 = df7.groupby("縣市").sum()
df7
```

```python
air = pd.read_csv("106-year.csv")
air = air.drop([0,1,2]) # drop first three useless rows
air
```

```python
air = air[['行政區','平均值']] # select only "行政區" & "平均值"
air
```

```python
df7_air = pd.merge(df7, air, left_on='縣市', right_on='行政區')
df7_air
```

## 到這裡下載 SHP 格式的「台灣村里界圖」
https://data.gov.tw/dataset/7440

```python
import numpy as np
import matplotlib.pyplot as plt

villages_shp = gp.read_file("mapdata201812120259/VILLAGE_MOI_121_1071130.shp") #全台灣村里界圖
taichung_shp = villages_shp.query('COUNTYNAME=="臺中市"') #臺中市村里界圖


#隨機產生臺中各區數值
taichung_shp['dummy number'] = np.random.ranf(taichung_shp.index.size)

taichung_shp.plot(
    cmap=plt.cm.Blues, #指定顏色
    column='dummy number' #指定從自身的這個 column 讀取顏色深度
)
plt.show()
```
