```python
import pandas as pd
```

```python
df = pd.read_csv("referendum_107.csv", thousands=',')
df
```

```python
# chose only proposal 7
df7 = df.loc[df['公投案'] == "第7案"]
df7
```

```python
df7 = df7.drop(columns=["公投案","鄉鎮市區","投票率(%)","有效同意票數對投票權人數百分比(%)"])
```

```python
df7 = df7.groupby("縣市").sum()
df7
```

```python
df7['有效同意票數對投票權人數百分比(%)'] = df7['同意票數']/df7['投票權人數']
df7
```

### 空氣品質資料

```python
air = pd.read_csv("106-year.csv")
air = air.drop([0,1,2]) # drop first three useless rows
air
```

```python
air = air[['行政區','平均值']] # select only "行政區" & "平均值"
air
```

```python
df7_air = pd.merge(df7, air, left_on='縣市', right_on='行政區')
df7_air
```

# 以 "(空氣品質指標)平均值" 對 "同意票數" 作散佈圖

```python
from pylab import mpl
mpl.rcParams['font.sans-serif'] = ['SimHei'] #將預設字體改用SimHei字體
```

```python
x = df7_air['平均值']
y = df7_air['有效同意票數對投票權人數百分比(%)']
n = df7_air['行政區']

fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(x, y)

for i, txt in enumerate(n):
    ax.annotate(txt, (x[i], y[i]))


plt.tight_layout()
```

```python
df7_air.plot(kind='scatter',title='空氣品質指標對同意票數',figsize=(8,6),x='平均值',y='有效同意票數對投票權人數百分比(%)',marker='o')
```
